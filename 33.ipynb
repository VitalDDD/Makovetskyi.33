```python
%matplotlib ipympl

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Читаємо файл з даними по авто
df_cars=pd.read_excel('car.xls')

# Дивимось перші записи даних по авто
df_cars.head()
```

```python
# Графік другого завдання - споживання від вартості авто лінійною діаграмою

fig, ax = plt.subplots()

plt.plot(df_cars['Price, USD'], df_cars['Fuel consumption, L/100 km'], color = 'green', linewidth = 2)
ax.set_xlabel("Price, USD")
ax.set_ylabel("Fuel consumption, L/100 km")
ax.set_title("Лінійна діаграма споживання палива від вартості авто")
```

```python
# Графік другого завдання - споживання від швидкості авто точковим графіком

fig, ax = plt.subplots()

plt.scatter(df_cars['Maximum speed, km/h'], df_cars['Fuel consumption, L/100 km'], color = 'red', marker = 'D')
ax.set_xlabel("Maximum speed, km/h")
ax.set_ylabel("Fuel consumption, L/100 km")
ax.set_title("Точковий графік споживання палива від швидкості авто")
```

```python
# Графік другого завдання - вантажопідйомність від швидкості авто точковим графіком

fig, ax = plt.subplots()

plt.scatter(df_cars['Maximum speed, km/h'], df_cars['Payload capacity, kg'], color = 'blue', marker = 'D')
ax.set_xlabel("Maximum speed, km/h")
ax.set_ylabel("Payload capacity, kg")
ax.set_title("Точковий графік вантажопідйомності від швидкості авто")
```

```python
# Третє завдання - гістограма розділення авто на 5 груп за швидкістю

fig, ax = plt.subplots()

plt.hist(df_cars['Maximum speed, km/h'], 6, color = 'red')
ax.set_xlabel("Maximum speed, km/h")
ax.set_ylabel("Number of cars")
ax.set_title("Гістограма розділення авто на 5 груп за швидкістю")
```

```python
# Четверте завдання - коробковий графік для відображення швидкостей різних марок
import seaborn as sns

fig, ax = plt.subplots()

sns.boxplot( y = 'Maximum speed, km/h', data = df_cars)

ax.set_xlabel("Car brand")
ax.set_ylabel("Maximum speed, km/h")
ax.set_title("Коробковий графік для відображення швидкостей різних марок")
```

```python
# П'яте завдання - парний графік для відображення залежностей величин

sns.pairplot(data = df_cars)
```
